```python
import pandas as pd
from evidently import Dataset, DataDefinition, Report
from evidently.descriptors import ExactMatch, RegExp, BeginsWith, EndsWith
 
# Load dataset
df = pd.read_csv("C:\\new\\health_and_fitness_qna.csv")
 
# Preview column names (adjust these if different)
print(df.columns)
 
# For the sake of the demo, create synthetic columns if needed
df["answer"] = df["Answer"].astype(str)
df["target"] = df["Answer"].astype(str)  
df["response"] = df["Question"].astype(str)
 
# Define which columns are text
data_def = DataDefinition(text_columns=["answer", "target", "response"])
 
# Add descriptors
eval_dataset = Dataset.from_pandas(
    df,
    data_definition=data_def,
    descriptors=[
        ExactMatch(columns=["answer", "target"], alias="exact_match"),
        RegExp(column_name="response", reg_exp=r"^How", alias="starts_with_How"),
        BeginsWith(column_name="response", prefix="What", case_sensitive=False, alias="begins_with_What"),
        EndsWith(column_name="response", suffix="?", case_sensitive=True, alias="ends_with_question_mark")
    ]
)
 
report = Report(metrics=[])
report.run(eval_dataset)
 
# Show the result
results = eval_dataset.as_dataframe()
print(results[[
    "answer", "target", "response",
    "exact_match", "starts_with_How", "begins_with_What", "ends_with_question_mark"
]])
 
 
```
